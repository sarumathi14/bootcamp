```python
import matplotlib.pyplot as plt
years = [1990, 1995, 2000, 2005, 2010, 2015, 2020]
phones_per_person = [0.02, 0.1, 0.3, 0.7, 1.0, 1.2, 1.5]
plt.figure(figsize=(8, 5))
plt.plot(phones_per_person, years, marker='o', linestyle='-', color='b', label='Phones per Person')
plt.xlabel('Number of Cell Phones per Person')
plt.ylabel('Year')
plt.title('Rise in Cell Phone Usage Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```
