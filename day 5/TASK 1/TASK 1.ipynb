```python
import re
import pandas as pd
commentary = [
    "Virat Kohli plays a stunning cover drive for four.",
    "Rohit Sharma pulls it over midwicket, that's a huge six!",
    "The bowler comes back strong with a perfect yorker.",
    "What a brilliant spell by Bumrah, three wickets already!",
    "MS Dhoni finishes in style with a helicopter shot for six.",
    "It's a nail-biting finish, the team needs 10 runs off the last over.",
    "A fantastic catch at the boundary saves crucial runs.",
    "The crowd is going wild as the batsman smashes another six!",
    "Jofra Archer delivers a 145 km/h yorker to Kohli!",
    "Bumrah bowls a 143 km/h inswinger, Rohit defends it."
]
bowler_pattern = r"(\b[A-Z][a-z]+\s[A-Z][a-z]+\b) (delivers|bowls)"
batter_pattern = r"to (\b[A-Z][a-z]+\b)"
ball_type_pattern = r"(yorker|bouncer|inswinger|outswinger|googly)"
shot_type_pattern = r"(cover drive|pull|hook|cut|single|boundary|six)"
speed_pattern = r"(\d{3})\s?km/h"
runs_pattern = r"(\d+)\s?(runs|run|four|six)"
extracted_data = []

for line in commentary:
    bowler = re.search(bowler_pattern, line)
    batter = re.search(batter_pattern, line)
    ball_type = re.search(ball_type_pattern, line)
    shot_type = re.search(shot_type_pattern, line)
    speed = re.search(speed_pattern, line)
    runs = re.search(runs_pattern, line)

    extracted_data.append({
        "Commentary": line,
        "Bowler": bowler.group(1) if bowler else None,
        "Batter": batter.group(1) if batter else None,
        "Ball Type": ball_type.group(1) if ball_type else None,
        "Shot Type": shot_type.group(1) if shot_type else None,
        "Speed (km/h)": int(speed.group(1)) if speed else None,
        "Runs Scored": int(runs.group(1)) if runs else None
         })
df_extracted = pd.DataFrame(extracted_data)
df_extracted
```
